# Demostración de Dendograma

```python
# Liberías para manipulación de datos
import pandas as pd
from pathlib import Path

# Librerías para crear el dendograma
from scipy.cluster.hierarchy import dendrogram, linkage

# Librerias para visualizar datos
import matplotlib.pyplot as plt
```

## Carga de Datos

```python
# Define la ruta del fichero
file_path = Path("../datos/demo_data_clustering.csv")

# Carga datos en un DataFrame
data = pd.read_csv(file_path)

# Datos de muestra
data.head()
```

## Clustering Jerárquico Aglomerativo

```python
# Realizar el clustering jerárquico usando SciPy
linked = linkage(data, 'ward')
```

## Visualización del Dendograma

```python
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Dendrograma')
plt.xlabel('Índice de Muestra')
plt.ylabel('Distancia')
plt.show()
```
