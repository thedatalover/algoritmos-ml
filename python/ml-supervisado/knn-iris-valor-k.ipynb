# Algoritmo KNN: Identificando el Mejor Valor para K

```python
# Librerías requeridas
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris # Importa el conjunto de datos Iris
from sklearn.model_selection import train_test_split # Para crear conjuntos de entrenamiento y prueba
from sklearn.neighbors import KNeighborsClassifier # Importa el algoritmo KNN
from sklearn.metrics import accuracy_score # Librerías para evaluación del modelo
```

```python
# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Probar diferentes valores de K para KNN y registrar la precisión
accuracy = []
k_range = range(1, 31)
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy.append(accuracy_score(y_test, y_pred))
```

```python
# Imprime valores de exactitud
print(accuracy)
```

```python
# Graficar la precisión en función de K
plt.figure(figsize=(12, 6))
plt.plot(k_range, accuracy, marker='o')
plt.title('Exactitud en función del valor de K para KNN')
plt.xlabel('Valor de K')
plt.ylabel('Exactitud')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```
